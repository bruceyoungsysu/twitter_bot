```python
path = './trump_tweets.txt'
with open(path) as f:
    lines = f.readlines()
```

```python
import nltk
from nltk.tokenize import word_tokenize
vocab = []
for line in lines:
    tokens = word_tokenize(line)
    vocab += tokens
```

```python
vocab_set = set(vocab)
```

```python
vocabs = [v for v in vocab_set if vocab.count(v)>100]
```

```python
with open('./vocab_tokenized_all.txt','w') as f:
    for item in vocab:
        f.write("%s\n" %item)
    f.close()
```
