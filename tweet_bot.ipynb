```python
import re
def get_vocab(file_path):
    vocab = []
    for line in open(file_path).readlines():
        vocab += [x.lower() for x in re.split('([.\?.,:!@#\n\s])', line)]
    vocab_set = set(vocab)
    vocab_count = [i for i in vocab_set if (i != ' ' and vocab.count(i)>100)]
    return vocab_count
```

```python
path = './trump_tweets.txt'
word_vocab = get_vocab(path)
```

```python
def encode_vocab(text, vocab):
    return [vocab.index(x)+1 for x in re.split('([.\?.,:!@#\n\s])', text) if x in vocab]
```

```python
with open('./trump_tweets.txt') as f:
    window = 10
    overlap = 5
    for line in f.readlines():

        text = encode_vocab(line, vocab)
        for start in range(0, len(text)-window, int(overlap)):
            chunk = text[start: start+window]
            chunk += [0]*(window-len(chunk))
            if len(chunk) != 10:
                print(chunk)
```
